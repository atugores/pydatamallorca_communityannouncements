```python
def binary_tree_ranking(n,k):
    assert n >= 2
    if n == 2:
        return ph.PhyloNetwork(eNewick="(1,2);")
    parent_tree = binary_tree_ranking(n-1, k / (2*(n-1)-1))
    node = parent_tree.nodes()[k % (2*(n-1)-1)]
    new_tree = ph.operations.push_and_hang(parent_tree,node,str(n))
    return new_tree
```

```python
for i in range(phgen.number_of_trees_bin_nont_global(4)):
    print binary_tree_ranking(4,i).eNewick()
```
